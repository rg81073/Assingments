### Q2. Can you explain the benefit of compound interest over a higher rate of interest that does not compound after reading this chapter?

Compound interest refers to the interest earned on both the principal amount and the accumulated interest on that principal amount. The key benefit of compound interest over a higher rate of interest that does not compound is the exponential growth of wealth over time. 

For instance, consider two investments: 

1. $10,000 at a fixed rate of 8% per annum (simple interest)
2. $10,000 at a fixed rate of 8% per annum (compounded annually)

After one year, the first investment will earn an interest of $800 (8% of $10,000). The total amount will be $10,800. 
On the other hand, the second investment will earn an interest of $800 (8% of $10,000) as well, but this amount will be added to the principal, making the total amount $10,800. 
For the second year, the first investment will earn an interest of $800 again, which will be added to the principal, making the total amount $11,600. But, the second investment will earn an interest of $864 (8% of $10,800), which will be added to the principal, making the total amount $11,665.60. 

As we can see, the second investment earned more interest compared to the first one even though the interest rate was the same. This is the power of compound interest. 

In Python, we can calculate the future value of an investment using compound interest using the following formula: 

```python
FV = P * (1 + r/n) ** (n*t)
```

where:
- FV: Future value of investment
- P: Principal amount
- r: Annual interest rate
- n: Number of times the interest is compounded per year
- t: Number of years

Here's an example of calculating the future value of an investment using compound interest in Python:

```python
P = 10000
r = 0.08
n = 1
t = 2

FV = P * (1 + r/n) ** (n*t)

print("Future value of investment:", FV)
```

Output:
```
Future value of investment: 11665.6
```

### Q4. If necessary, how do you change the aspect ratios between the X and Y axes? 

To change the aspect ratios between the X and Y axes, you can use the aspect parameter of the matplotlib.pyplot module's subplots() function or axis() function.

Here is an example of using the aspect parameter with subplots() function:


````python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data
x = np.random.rand(100)
y = np.random.rand(100)

# Creating a scatter plot with equal aspect ratio
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_aspect('equal')

# Creating a scatter plot with different aspect ratio
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_aspect(2.0)
````

In this example, the set_aspect() function is used to set the aspect ratio of the scatter plot. The first scatter plot has an equal aspect ratio, whereas the second has a different aspect ratio of 2.0.

You can also use the aspect parameter with the axis() function to set the aspect ratio of the plot. For example:

````python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data
x = np.random.rand(100)
y = np.random.rand(100)

# Creating a scatter plot with different aspect ratio
plt.scatter(x, y)
plt.axis('equal')
plt.show()

# Creating a scatter plot with different aspect ratio
plt.scatter(x, y)
plt.axes().set_aspect(2.0)
plt.show()
````

In this example, the first scatter plot has an equal aspect ratio, whereas the second has a different aspect ratio of 2.0, which is set using the set_aspect() function of the axes() object.

### Q5. Compare and contrast the three types of array multiplication between two numpy arrays: dot product, outer product, and regular multiplication of two numpy arrays.

In numpy, there are three types of multiplication between two arrays: dot product, outer product, and element-wise multiplication.

1. Dot Product: Dot product is also called matrix multiplication, where the dot() function is used to compute the dot product of two arrays. The dot product between two arrays of rank 1 results in scalar output, while dot product between arrays of rank greater than 1 results in an array of rank N-2. 

Example:

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.dot(a, b)
print(c)
```

Output:
```
32
```

2. Outer Product: The outer product between two arrays can be obtained using the outer() function. It results in an array of rank 2, whose shape is equal to the product of the shapes of the input arrays.

Example:

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.outer(a, b)
print(c)
```

Output:
```
[[ 4  5  6]
 [ 8 10 12]
 [12 15 18]]
```

3. Element-wise multiplication: The element-wise multiplication is done using the * operator. The shape of the resulting array is the same as the input arrays.

Example:

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = a * b
print(c)
```

Output:
```
[ 4 10 18]
```

### Q6. Before you buy a home, which numpy function will you use to measure your monthly mortgage payment?

NumPy provides a function called `PMT` (Payment) to calculate the monthly mortgage payment. 

The formula to calculate monthly mortgage payment is:

`M = P [ r(1 + r)^n / ((1 + r)^n  1)]`

where:
- `M` is the monthly payment
- `P` is the principal amount
- `r` is the interest rate
- `n` is the number of payments

Using the `PMT` function, we can calculate the monthly mortgage payment as follows:

``` python
import numpy as np

# principal amount
P = 100000

# interest rate
r = 0.05

# number of payments
n = 360

# calculate monthly mortgage payment
monthly_payment = np.pmt(rate=r/12, nper=n, pv=-P)

print("Monthly Mortgage Payment: $", round(monthly_payment, 2))
```

Output:
```
Monthly Mortgage Payment: $ 536.82
```

So, the monthly mortgage payment for a principal amount of $100,000, an interest rate of 5%, and a repayment period of 30 years (360 months) would be $536.82.

### Q7. Can string data be stored in numpy arrays? If so, list at least one restriction that applies to this data. 

Yes, string data can be stored in numpy arrays. However, there are some restrictions that apply:

1. All strings in a numpy array must have the same length.
2. The length of the strings in a numpy array is fixed when the array is created, so it cannot be changed later.
3. String arrays can be slower than other arrays in numpy due to their variable length.

Here's an example of creating a numpy array of strings:

```python
import numpy as np

# Create a numpy array of strings with fixed length of 10 characters
str_array = np.array(['apple', 'banana', 'cherry'], dtype='S10')

# Access elements in the array
print(str_array[0])  # b'apple'
print(str_array[1])  # b'banana'
print(str_array[2])  # b'cherry'
```

In this example, the `dtype` parameter is used to specify that the array should be of type string (`'S'`) and that each string should have a length of 10 characters (`'S10'`). The prefix `b` indicates that the strings are represented as bytes in the array.
